```python
import pandas as pd

df=pd.read_excel(r"C:\New Folder\Sample - Superstore.xls\Sample - Superstore.xls")

print(df.head())
```

```python
category_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
category_sales
```

```python
plt.figure()
category_sales.plot(kind='bar')
plt.title('Category-wise Sales')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.show()
```

```python
top_products=(
    df.groupby('Sub-Category')['Sales']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)
top_products
```

```python
plt.figure()
top_products.plot(kind='bar')
plt.title('Top 10 Products by Sales')
plt.xlabel('Sub-Category')
plt.ylabel('total Sales')
plt.show()
```

```python
profit_by_category=df.groupby('Category')['Profit'].sum()
profit_by_category
```

```python
plt.figure()
profit_by_category.plot(kind='bar')
plt.title('Profit by Category')
plt.xlabel('Category')
plt.ylabel('Profit')
plt.show()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
```

```python
df[['Order Date','Year','Month']].head()
```

```python
monthly_trend=(
    df.groupby(['Year','Month'])[['Sales','Profit']]
    .sum()
    .reset_index()
)
monthly_trend.head()
```

```python
plt.figure()
for year in monthly_trend['Year'].unique():
    yearly_data=monthly_trend[monthly_trend['Year']==year]
    plt.plot(yearly_data['Month'],yearly_data['Sales'],label=year)
plt.title('Monthly Sales Trend (year-wise)')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend()
plt.show()
```
